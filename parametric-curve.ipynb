```python
#plot the image of a circle lattice under the mobius transformation z --> 1/z
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [10,10]
plt.rcParams["figure.autolayout"] = True

N = 1000
M = 30
r = 0.2

t = np.linspace(0, 2 * np.pi, N)
lx = []
ly = []

for i in range(0,M):
    for j in range(0,M):
      lx.append(np.array(r * np.cos(t) + 2*r*i))
      ly.append(np.array(r * np.sin(t) + 2*r*j))

x = []
y = []

for i in range(0,M):
    for j in range(0,M):
      x.append(np.array([lx[i][j]/(lx[i][j]**2 + ly[i][j]**2) for j in range(0, len(t))]))
      y.append(np.array([-ly[i][j]/(lx[i][j]**2 + ly[i][j]**2) for j in range(0, len(t))]))

fig, ax = plt.subplots()
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)

ax.set_aspect('equal')
  
    
for i in range(0,M*M):
  ax.plot(lx[i], ly[i])
  ax.plot(x[i], y[i])
```

```python
#plot the image of horizontal lines under a mobius transformation
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = [10,10]
plt.rcParams["figure.autolayout"] = True

N = 1000
M = 30
r = 0.2

#compute real part of transformation output
def rpart(x, y):
    num = 4*(x**2+y**2)-(5*x)+4*(x-1)
    denom = (2*x+2)**2+(2*y-1)**2
    return num/denom

#compute imaginary part of transformation output
def ipart(x, y):
    num = (2*x**2+2*y**2+5*y-4*x*y-2)
    denom = (2*x+2)**2+(2*y-1)**2
    return num/denom

t = np.linspace(0, 2 * np.pi, N)
lx = []
ly = []

for i in range(0,M):
    for j in range(0,M):
      lx.append(np.array(r * np.cos(t) + 2*r*i))
      ly.append(np.array(r * np.sin(t) + 2*r*j))

x = []
y = []

for i in range(0,M):
    for j in range(0,M):
      x.append(np.array([rpart(lx[i][j], ly[i][j]) for j in range(0, len(t))]))
      y.append(np.array([ipart(lx[i][j], ly[i][j]) for j in range(0, len(t))]))

fig, ax = plt.subplots()
ax.set_xlim(-2,2)
ax.set_ylim(-5,5)

ax.set_aspect('equal')
  
    
for i in range(0,M*M):
  #ax.plot(lx[i], ly[i])
  ax.plot(x[i], y[i])
```

```python
#plot the image of a lattice under a mobius transformation using python complex numbers
import numpy as np
import matplotlib.pyplot as plt
import cmath

plt.rcParams["figure.figsize"] = [10,10]
plt.rcParams["figure.autolayout"] = True

N = 60
M = 30

def ctransform(x):
    return x

t = np.linspace(-10,10, N)
#l holds the vertical line coordinates
l = []

for i in range(-M,M):
    for j in t:
      l.append(complex(i,j))

#m holds the horizontal line coordinates
m = []

for i in range(-M,M):
    for j in t:
      m.append(complex(j,i))

#ox, oy are the coordinates of the points to plot
ox = []
oy = []

#transform vertical lines
for i in range(0,len(l)):
    w = ctransform(l[i])
    ox.append(w.real)
    oy.append(w.imag)

#transform horizontal lines
for i in range(0,len(m)):
    w = ctransform(m[i])
    ox.append(w.real)
    oy.append(w.imag)

fig, ax = plt.subplots()
#x.set_xlim(-100,100)
ax.set_ylim(-100,100)

#ax.set_aspect('equal')

for i in range(0, len(ox)):
  ax.plot(ox[i],oy[i])

plt.show()
```

```python
#plot the image of a circle lattice under a mobius transformation using python complex numbers
import numpy as np
import matplotlib.pyplot as plt
import cmath

plt.rcParams["figure.figsize"] = [6.5,6.5]
plt.rcParams["figure.autolayout"] = True

#number of points in a circle
N = 1000
#side half-length of circle lattice
M = 40
#radius of circle
R = 1

#mobius transformation
def ctransform(x):
    return 10/x 

#create the circle lattice
t = np.linspace(0, 2 * np.pi, N)
circx = np.array([R * np.cos(x) for x in t])
circy = np.array([R * np.sin(x) for x in t])
circlesx = []
circlesy = []

for h in range(-M, M):
  for k in range(-M, M):
    circlesx.append(circx + 2*R*h)
    circlesy.append(circy + 2*R*k)

tx2 = np.array(circlesx)
ty2 = np.array(circlesy)
c = tx2 + 1j * ty2

ox = []
oy = []
#ox, oy are the coordinates of the points to plot
for i in range(0,len(c)):
    w = ctransform(c[i])
    ox.append(w.real)
    oy.append(w.imag)

fig, ax = plt.subplots()
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)

for i in range(0, len(ox)):
  ax.plot(ox[i],oy[i])

plt.show()
#plt.savefig('mobius.jpg')
```

```python
#plot the image of a line lattice under a mobius transformation using python complex numbers
import numpy as np
import matplotlib.pyplot as plt
import cmath

plt.rcParams["figure.figsize"] = [6.5,6.5]
plt.rcParams["figure.autolayout"] = True

#number of points in a line -
#as this is made large,the result approaches a transformed circle lattice
N = 18
#side half-length of line lattice
M = 10

#mobius transformation
def ctransform(x):
    return 20/x

#create the circle lattice
# t is the line index
t = np.linspace(-M, M, N)
#t1 fills in with points each line
t1 = np.linspace(-M, M, 2*M+1)

#cartesian coordinates of lines
vlinesx = []
vlinesy = []
hlinesx = []
hlinesy = []

#complex coordinates of lines
vlines = []
hlines = []

#transformed lines (output)
ox = []
oy = []

#create the vertical lines
for h in t1:
    vlinesx.append(np.full((1,len(t)), h).ravel())
    vlinesy.append(t)
                   
#create the horizontal lines
for k in t1:
    hlinesx.append(t)              
    hlinesy.append(np.full((1,len(t)), k).ravel())

#change list to numpy to form complex numbers
txv = np.array(vlinesx)
tyv = np.array(vlinesy)

txh = np.array(hlinesx)
tyh = np.array(hlinesy)

#convert cartesian coordinates of lines to complex coordinates
for i in range(0,len(txv)):
  vlines.append(txv[i] + 1j * tyv[i])

for j in range(0,len(txh)):
  hlines.append(txh[j] + 1j * tyh[j])

#transform the lines
for i in range(0,len(vlines)):
    w = ctransform(vlines[i])
    ox.append(w.real)
    oy.append(w.imag)
              
for j in range(0,len(hlines)):
    w = ctransform(hlines[j])
    ox.append(w.real)
    oy.append(w.imag)

fig, ax = plt.subplots()
ax.set_xlim(-10,10)
ax.set_ylim(-10,10)

for i in range(0, len(ox)):
  ax.plot(ox[i],oy[i])

plt.show()
```

```python
#plot the image of a set of concentric circles under a mobius transformation using python complex numbers
import numpy as np
import matplotlib.pyplot as plt
import cmath

plt.rcParams["figure.figsize"] = [6.5,6.5]
plt.rcParams["figure.autolayout"] = True

#number of points in an object
N = 1000
#radius of circle
R = 1.2
#number of circles
M = 16

#mobius transformation
def ctransform(x):
    return 40/x

t = np.linspace(0, 2 * np.pi, N)

#cartesian coordinates of concentric circles
x = []
y = []

for i in range(0,M):
    x.append((0.8)**i * R * (np.cos(t) + np.cos(6*t)/2 + np.sin(14*t)/6))
    y.append((0.8)**i * R * (np.sin(t) + np.sin(6*t)/2 + np.cos(14*t)/6))

#complex coordinates of the input circles
cinput = []

#cartesian coordinates of the output
ox = []
oy = []

#convert cartesian coordinates of circles to complex coordinates
for i in range(0, len(x)):
  cinput.append(x[i] + 1j * y[i])

#transform the circles
for i in range(0,len(cinput)):
    w = ctransform(cinput[i])
    ox.append(w.real)
    oy.append(w.imag)

#plot the transformed concentric circles
fig, ax = plt.subplots()
ax.set_xlim(-3000,3000)
ax.set_ylim(-3000, 3000)
ax.set_facecolor('black')

for i in range(0, len(ox)):
    ax.plot(ox[i],oy[i])


plt.show()
```
